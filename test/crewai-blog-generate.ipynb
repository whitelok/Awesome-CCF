```python
from crewai import Agent, Task, Crew
from langchain_openai import ChatOpenAI
import os

openai_gpt4_llm = ChatOpenAI(base_url="https://xy.eveai.top/v1", 
                 model="gpt-4-turbo-preview", 
                 api_key="YplMYpwI6M40MRVH7dB5Ac2f2c6c434dB405B3860457B29f")
```

```python
planner = Agent(
    role="内容策划",
    goal="根据{topic}策划引人入胜且事实准确的内容。",
    backstory="你正在策划发表一篇关于{topic}的博客文章。"
              "你需要收集可以帮助读者了解关于{topic}的细节和信息。"
              "您必须准备一份详细的大纲以及博客文章中必须包含的相关主题和子主题。"
              "将输出内容为中文，如果不是中文，请根据上下文和相关知识翻译成中文，相关术语或关键词可以保留英文。",
    llm=openai_gpt4_llm,
    allow_delegation=False,
    verbose=True
)
```

```python
writer = Agent(
    role="内容撰写",
    goal="针对{topic}撰写深刻且事实准确的博客文章。",
    backstory="你正在给'https://zhihu.com/'撰写一篇新的关于{topic}的评论。"
              "根据内容策划给出的大纲和主题，详细撰写内容。"
              "如果章节讨论的是某个技术框架，应该添加对该技术框架的代码实例。",
    allow_delegation=False,
    llm=openai_gpt4_llm,
    verbose=True
)
```

```python
editor = Agent(
    role="编辑",
    goal="编辑给定博客文章，使其符合'https://zhihu.com/'的写作风格。",
    backstory="你作为编辑接收内容撰写发给你的博客文章。"
              "您的目标是审查博客文章，确保它遵循新闻业的最佳实践，提供出中肯的意见和观点。"
              "并且尽可能避免涉及争议性话题或意见。",
    llm=openai_gpt4_llm,
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. 优先关注{topic}上的最新趋势、关键参与者和值得注意的新闻。\n"
        "2. 确定目标受众，关注他们的兴趣和痛点。\n"
        "3. 制定一个详细的内容大纲，包括引言、主要观点和行动号召。\n"
        "4. 包含SEO关键词以及相关数据或来源。\n"
        "5. 将输出内容为中文，如果不是中文，请根据上下文和相关知识翻译成中文，相关术语或关键词可以保留英文。"
    ),
    expected_output="一份全面的内容策划文档，包括大纲、受众分析、SEO关键词和相关素材。",
    agent=planner,
)
```

```python
write = Task(
    description=(
        "1. 利用内容策划撰写关于{topic}的引人入胜的博客文章。\n"
        "2. 自然地将SEO关键词融入到文章中。\n"
        "3. 各部分/小标题应以吸引人的方式恰当命名。\n"
        "4. 确保文章结构包括一个吸引人的引言、内容丰富的主体和总结性的结论。\n"
        "5. 校对文章，确保没有语法错误。\n"
        "6. 将输出内容为中文，如果不是中文，请根据上下文和相关知识翻译成中文，相关术语或关键词可以保留英文。\n"
        "7. 文章的每个章节应该有至少2个段落。每个章节的文字不少于1000字。"
    ),
    expected_output="一篇精彩的markdown格式的博客文章。"
        "文章完整度已经可以达到随时可以发表的水平。"
        "文章的每个章节应该有2至3个段落。"
        "每个章节的文字不少于1000字。"
        "文章内容主要是中文，相关术语或关键词可以保留英文。",
    agent=writer,
)
```

```python
edit = Task(
    description=("1. 校对给定的博客文章，检查其语法错误，统一全文内容风格。\n"
                 "2. 将输出内容为中文，如果不是中文，请根据上下文和相关知识翻译成中文，相关术语或关键词可以保留英文。"),
    expected_output="一篇精彩的markdown格式的博客文章。"
                    "文章完整度已经可以达到随时可以发表的水平。"
                    "文章的每个章节应该有2至3个段落。"
                    "文章内容主要是中文，相关术语或关键词可以保留英文。",
    agent=editor
)
```

```python
# crew = Crew(
#     agents=[planner, writer, editor],
#     tasks=[plan, write, edit],
#     verbose=2
# )

crew = Crew(
    agents=[planner, writer],
    tasks=[plan, write],
    verbose=2
)
```

```python
inputs = {"topic":"LangGraph使用实践"}
result = crew.kickoff(inputs=inputs)
```
